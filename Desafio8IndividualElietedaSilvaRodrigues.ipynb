```python
url = "https://raw.githubusercontent.com/dsreliete/Womakerscode_DA_desafio_individual/refs/heads/main/vendas.csv"
df = pd.read_csv(url)
```

```python
# 1. Carregar os dados de um arquivo CSV fornecido (vendas.csv).
df
```

```python
# 2. Criar um gráfico de linhas das vendas ao longo do tempo usando Matplotlib.

df['Date'] = pd.to_datetime(df['date'])
plt.plot(df['date'], df['sales'])
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Linha de vendas ao longo do tempo')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Criar um gráfico de decomposição sazonal usando Seaborn.

df['Date'] = pd.to_datetime(df['date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)
decomposition = sm.tsa.seasonal_decompose(df['sales'], model='additive')
fig = decomposition.plot()
fig.set_size_inches(15, 8)
plt.show()
```

```python
# 4. Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly.

df['Date'] = pd.to_datetime(df['date'])
fig = px.line(df, x='date', y='sales', title='Linha de vendas ao longo do tempo')
fig.update_xaxes(title_text='Date')
fig.update_yaxes(title_text='Sales')
fig.show()
```
